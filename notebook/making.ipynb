```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 500)
```

```python
train_series = pd.read_parquet(
        "/kaggle/input/preprocessed_train_series_le_50_fold.parquet"
    )
print("series_df loaded.")
```

```python
train_series.columns
```

```python
# series_date_keyごとのeventの-1の数をカウント
train_series["is_unknown"] = (train_series["event"] == -1).astype(int)
train_series["pseudo_count"] = train_series.groupby("series_date_key")["is_unknown"].transform("sum")
```

```python
# train_series["pseudo_count"] = train_series.groupby("series_date_key")[
#         "event"
# ].count()
display(train_series[["series_date_key", "event", "pseudo_count"]].head(10))
```

```python
train_series["is_pseudo_target"] = (train_series["pseudo_count"] > 0).astype(
    "int8"
)
display(train_series[["series_date_key", "event", "is_pseudo_target"]].head())
```

```python
# series_date_keyごとのpseudo_targetが0より大きいもののseries_date_keyを取得
pseudo_target_key_list = train_series[
    train_series["is_pseudo_target"] > 0
]["series_date_key"].unique()
```
